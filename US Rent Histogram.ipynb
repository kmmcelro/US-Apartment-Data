```python
"""
Author: Kaitlin McElroy
Data Set taken from ApartmentList's website for September 2022
Current code: takes data from an excel file and sorts it by how many bedrooms that number describes. 
The value is saved to an array if the data is not NA and made into a histogram
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import openpyxl

dataframe = openpyxl.load_workbook('Apartment_List_Rent_Estimates_City_2022_9.xlsx')
dataframe2 = dataframe.active # points to the sheet that was active during the last save

sheetnm = dataframe.sheetnames[0] # gets the name of the first sheet on the excel file
dataframe1 = dataframe[sheetnm] # selects it to work with

data = np.array([]) # defining a NumPy array for the overall data for each city available

for j in range(2,1729): # loops over the rows in the data set to add to array - could combine this with below loop
    cold = "D" + str(j)
    cole = "E" + str(j)
    if dataframe1[cold].value == "_Overall":
        if dataframe1[cole].value != '     NA': # unnessary since the Overall data has a number for each city available
            data = np.append(data, dataframe1[cole].value)

ave = np.average(data) # calculates the average value of the data set
Ave = round(ave,3)
Ave = str(Ave)

Std = np.std(data) # calucates the standard deviation 
STD = round(Std,3)
STD = str(STD)

# drawing the line for average value and ~68% interval
plt.axvline(x=ave,linewidth = 1, color = "red")
plt.axvline(x=ave+Std,linewidth = 1, color = "blue")
plt.axvline(x=ave-Std,linewidth = 1, color = "blue")

# adding the histogram and graph labels
plt.title("Distribution of Overall Average Rent Prices in US Cities\nFall 2022",fontsize=10)
plt.xlabel("Overall Average for All Cities= "+Ave+r"$ \pm $"+STD+" \n",fontsize=10)
plt.hist(data,bins = 100, color = 'purple')
```

```python
# since overall could be skewed by number of bedrooms, let's see the one bedroom specific data
# doing the same routine as used for Overall, but column d is checked for 1br instead of _Overall
data1 = np.array([])
for j in range(2,1729):
    cold = "D" + str(j)
    cole = "E" + str(j)
    if dataframe1[cold].value == "1br":
        if dataframe1[cole].value != '     NA':
            data1 = np.append(data1, dataframe1[cole].value)

# Calculating average and standard deviation for this new data set
ave = np.average(data1)
Ave = round(ave,3)
Ave = str(Ave)
Std = np.std(data1)
STD = round(Std,3)
STD = str(STD)

# Graphing our histogram with the lines indicating average and the ~68% range (1 std)
plt.axvline(x=ave,linewidth = 1, color = "red")
plt.axvline(x=ave+Std,linewidth = 1, color = "blue")
plt.axvline(x=ave-Std,linewidth = 1, color = "blue")
plt.title("Distribution of Average Rent Prices for 1 Bedroom Rentals in US Cities\nFall 2022",fontsize=10)
plt.xlabel("Overall Average for 1 Bedroom Rentals for All Cities= "+Ave+r"$ \pm $"+STD+" \n",fontsize=10)
plt.hist(data1,bins = 100, color = 'purple')
```

```python
# plotting the population of a city vs average cost of 1 bed rental there 
POP = np.array([]) # getting population of every city with 1 bedroom rental data
for j in range(2,1729):
    cold = "D" + str(j)
    colc = "C" + str(j)
    cole = "E" + str(j)
    if dataframe1[cold].value == "1br":
        if dataframe1[cole].value != '     NA':
            POP = np.append(POP, dataframe1[colc].value)

#plt.ylabel('Average Cost of 1 Bedroom Apartment')
#plt.xlabel('Population of City')
#plt.scatter(POP,data1)

plt.ylabel('Average Cost of 1 Bedroom Apartment') # plotting on linear scale to take initial look at data pair distribution
plt.xlabel('Population of City')
plt.plot(POP, data1,"bo") 
plt.show()

plt.ylabel('Average Cost of 1 Bedroom Apartment') # plotting log vs log to make any relation between parameters more apparent
plt.xlabel('Population of City')
plt.loglog(POP, data1,"bo") 
```
